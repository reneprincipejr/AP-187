```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
#white = np.loadtxt("white")
```

```python
DGK_wav, DGK_int = np.loadtxt("dgk_wav.txt"), np.loadtxt("DGK_intensity.txt")
```

```python
wavelength = np.arange(400,700,0.5)
intensities = np.zeros(shape=[len(wavelength), np.shape(DGK_int)[1]])
for i in range(np.shape(DGK_int)[1]):
    intensities[:,i] = savgol_filter(np.interp(wavelength, DGK_wav, DGK_int[:,i]), 11,2)
    
white = intensities[:,0]

reflectances = np.zeros(shape=[len(wavelength), np.shape(DGK_int)[1]-1])
for i in range(np.shape(DGK_int)[1]-1):
    reflectances[:,i] = intensities[:,i+1]/white

for i in range(14):
     reflectances[:,i] = reflectances[:,i]/reflectances[:,-2]
    
#wavelength = np.arange(400,700,1)

'Sensitivity'

red = np.loadtxt("red.txt")
green = np.loadtxt("green.txt")
blue = np.loadtxt("blue.txt")

#wavelength = np.arange(400,700,.5)

r = savgol_filter(np.interp(wavelength, red[:,0], red[:,1]), 11, 2)
g = savgol_filter(np.interp(wavelength, green[:,0], green[:,1]), 11, 2)
b = savgol_filter(np.interp(wavelength, blue[:,0], blue[:,1]), 11, 2)

red_c = np.loadtxt("CAVE_RED.txt")
green_c = np.loadtxt("CAVE_GREEN.txt")
blue_c = np.loadtxt("CAVE_BLUE.txt")

#wavelength = np.arange(400,700,.5)

r_c = savgol_filter(np.interp(wavelength, red_c[:,0], red_c[:,1]), 11, 2)
g_c = savgol_filter(np.interp(wavelength, green_c[:,0], green_c[:,1]), 11, 2)
b_c = savgol_filter(np.interp(wavelength, blue_c[:,0], blue_c[:,1]), 11, 2)


'Loading Macbeth dataset'

macbeth = np.loadtxt("macbeth.txt")
macbeth_wav, macbeth_ref =  macbeth[4:64,0], macbeth[4:64,1:]
macbeth_r = np.zeros(shape=[len(wavelength),24])


for i in range(1,25):
    macbeth_r[:,i-1] = np.interp(wavelength, macbeth_wav, macbeth_ref[:,i-1])
    
'Radiance'

light = np.loadtxt("LED.txt")
source = np.interp(wavelength, light[:,0], light[:,1])
```

```python
intensities = np.zeros(shape=[len(wavelength), np.shape(DGK_int)[1]])
for i in range(np.shape(DGK_int)[1]):
    intensities[:,i] = savgol_filter(np.interp(wavelength, DGK_wav, DGK_int[:,i]), 21,2)
    
white = intensities[:,0]

reflectances = np.zeros(shape=[len(wavelength), np.shape(DGK_int)[1]-1])
for i in range(np.shape(DGK_int)[1]-1):
    reflectances[:,i] = intensities[:,i+1]/white

for i in range(14):
     reflectances[:,i] = reflectances[:,i]-reflectances[:,-3]
    

plt.plot(reflectances)
plt.plot(white, 'k')
plt.show()
```

```python
def Color_Simulation(r_sensitivity,g_sensitivity,b_sensitivity, source, kr, kg, kb, transmittance):

    plt.figure(figsize=[30,18])
    plt.rcParams['axes.titlesize'] = 20
    plt.rcParams.update({'font.size': 18})
    plt.rcParams['figure.facecolor'] = 'w'

    R_sim = []
    G_sim = []
    B_sim = []
    
    for i in range(0,24):
        plt.subplot(4,6,i+1)
        R = np.trapz(macbeth_r[:,i]*r_sensitivity*source*transmittance)
        G = np.trapz(macbeth_r[:,i]*g_sensitivity*source*transmittance)
        B = np.trapz(macbeth_r[:,i]*b_sensitivity*source*transmittance)
        P = np.empty(shape=[15,15,3])
        
        R_sim.append(R*kr)
        G_sim.append(G*kg)
        B_sim.append(B*kb)
        
        for x in range(15):
            for y in range(15):
                P[x,y] = [R*kr,G*kg,B*kb]
        plt.title(str(i+1))
        plt.imshow(P)
        plt.xlabel('RGB = [' + str(round(R*kr*255,0)) + 
                  ', ' + str(round(G*kg*255,0)) + ', '
                  + str(round(B*kb*255,0)) + ']')
        plt.yticks([])
        plt.xticks([])
        
    return R_sim, G_sim, B_sim
#R_sim, G_sim, B_sim = Color_Simulation(r,g,b,source, 1/45.8, 1/70, 1/42.7, 1)
R_sim, G_sim, B_sim = Color_Simulation(r,g,b,1, 1/185, 1/124, 1/119, 1)
#R_sim, G_sim, B_sim = Color_Simulation(r_c,g_c,b_c,source, 1/43.5, 1/72.2, 1/41.7, 1)

```
