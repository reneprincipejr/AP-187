```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants
from scipy.constants import h,c,k
from mpl_toolkits.axes_grid.inset_locator import inset_axes
from scipy.signal import savgol_filter
```

```python
cie = np.loadtxt("CIE1964.txt")
```

```python
wavelength, x_, y_, z_ = cie[:,0], cie[:,1], cie[:,2], cie[:,3]
```

```python
def Tongue(x_,y_,z_):
    bound_x, bound_y = [],[]
    for i in range(len(x_)):
        X,Y,Z = x_[i], y_[i], z_[i]
        x = X/(X+Y+Z)
        y = Y/(X+Y+Z)
        bound_x.append(x)
        bound_y.append(y)
        
    return bound_x, bound_y
        
```

```python
boundary_x,boundary_y = Tongue(x_,y_,z_)
```

```python
fig = plt.figure(figsize=[15,8], facecolor = 'white')
plt.rcParams['axes.titlesize'] = 20
plt.rcParams.update({'font.size': 20})

wav = np.arange(200e-9,2000e-9,10e-9)
Temp = np.arange(1000,10000,1000)

blackbody_spectrum = np.zeros(shape=[len(wavelength),len(Temp)])

for T in Temp:
    def P(wav, T_max,N):
        a = N/(18*((wav**5)*(np.exp((h*c)/(wav*k*T))-1)))
        return a, np.trapz(a)

    b, b_total = P(wav, T, 2e-28)
    wav1 = np.arange(200,2000,10)
    
    #blackbody_spectrum[:,Temp.tolist().index(T)] = b/b_total
    spectrum = np.interp(wavelength, wav1, b/b_total)
    blackbody_spectrum[:,Temp.tolist().index(T)] = spectrum
    
    plt.plot(wav1, b/b_total, '-', label = str(T) + 'K')
    plt.xlabel('Wavelength, $\lambda$, nm')
    plt.ylabel('Power/Wavelength interval, $W/nm$')
    plt.legend(loc=1)
    plt.xticks(np.arange(200,2001,200))
plt.grid()
```

```python
def Locus(spectrum, x_, y_, z_):
    locus_x, locus_y = [],[]
    for T in range(spectrum.shape[1]):
        X,Y,Z = np.trapz(spectrum[:,T]*x_), np.trapz(spectrum[:,T]*y_), np.trapz(spectrum[:,T]*z_)
        x = X/(X+Y+Z)
        y = Y/(X+Y+Z)
        locus_x.append(x)
        locus_y.append(y)
        
    return locus_x, locus_y
    
```

```python
locus_x, locus_y = Locus(blackbody_spectrum, x_,y_,z_)
```
